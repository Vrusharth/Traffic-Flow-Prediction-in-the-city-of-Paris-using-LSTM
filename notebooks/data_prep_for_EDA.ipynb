## Accessing raw data

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_parquet('../data/vehicles_EDA/[Paris] Amsterdam x Clichy.parquet')

df['Hour'] = df['t'].dt.hour
df['Day of the Month'] = df['t'].dt.day
df['Day Name'] =  df['t'].dt.day_name()
df = df.reset_index(drop=True)


df_hour = df.groupby('Hour')[['2_wheelers', '4_wheelers']].mean().reset_index()
df_hour['2_wheelers'] = df_hour['2_wheelers'].astype(int)
df_hour['4_wheelers'] = df_hour['4_wheelers'].astype(int)


df_day = df.groupby('Day of the Month')[['2_wheelers', '4_wheelers']].mean().reset_index()
df_day['2_wheelers'] = df_day['2_wheelers'].astype(int)
df_day['4_wheelers'] = df_day['4_wheelers'].astype(int)


df_day_name = df.groupby('Day Name')[['2_wheelers', '4_wheelers']].mean().reset_index()
df_day_name['2_wheelers'] = df_day_name['2_wheelers'].astype(int)
df_day_name['4_wheelers'] = df_day_name['4_wheelers'].astype(int)
```

```python
df_day_name
```

```python
cats = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_day_name.groupby(['Day Name']).sum().reindex(cats).reset_index()
```

```python
import plotly.express as px
fig = px.line(df_day_name, x="Day Name", y=['2_wheelers', '4_wheelers'], labels={'variable' : 'type of vehicle', 'value' : 'Vehicles',
                                                                         'hour' : 'Hour'},
#              hover_data={"date": "|%B %d, %Y"},
              title='Average Traffic by day of month', markers=True)
fig.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 0.0,
        dtick = 1.0
    )
)
fig.update_traces(marker_size=10)
fig.show()
```
